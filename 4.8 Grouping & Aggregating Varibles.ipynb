## 2. Aggregating Mean

```python
# aggregating mean/groupby

ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

## 4.Transform and Loc

```python
# creating max_orders column to create loyalty flag

ords_prods_merge['max_orders'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head()
```

```python
# creating loyalty flag

ords_prods_merge.loc[ords_prods_merge['max_orders'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_orders'] <= 40) & (ords_prods_merge['max_orders'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_orders'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

### There's no real difference of the types of products that are purchased between loyal, regular, and new customers. It only differs between the amount of items that are purchased.

## 6. Spender Flag

```python
# creating mean_order column to create spender flag

ords_prods_merge['mean_order'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
#creating spending flag/low spender

ords_prods_merge.loc[ords_prods_merge['mean_order'] < 10, 'spender_flag'] = 'Low spender'
```

```python
#creating spending flag/high spender

ords_prods_merge.loc[ords_prods_merge['mean_order'] >= 10, 'spender_flag'] = 'High spender'
```

```python
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

## 7. Order Frequency Flag

```python
# creating frequency column to create a flag

ords_prods_merge['frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#creating order frequency flag/non-frequent customer

ords_prods_merge.loc[ords_prods_merge['frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
#creating order frequency flag/regular customer

ords_prods_merge.loc[(ords_prods_merge['frequency'] > 10) & (ords_prods_merge['frequency'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
#creating order frequency flag/frequent customer

ords_prods_merge.loc[ords_prods_merge['frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
